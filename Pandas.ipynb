**Merge**
1. It acts like a SQL joins where we coonect the dataframes based on common columns or(keys)

```python
students=pd.DataFrame({
    "ID":[11,22,33,44],
    "Name":["Mahesh","Kumar","Mahi","Mahesh Kumar"]
})
students
```

```python
scores=pd.DataFrame({
    "ID":[2,4,6,8],
    "Scores":[98,78,67,56]
})
scores
```

**We will use basic merge(Inner Join in SQL)**

here we will get the IDs presentin both table

```python
merged=pd.merge(students,scores,on="ID")
merged
```

**Left Join**

It keeps all the rows from left table(Student),and put null values if it is not present in Right Table(Scores)

```python
leftMerged=pd.merge(students,scores,on="ID",how="left")
leftMerged
```

```python
leftMerged=pd.merge(students,scores,on="ID",how="right")
leftMerged
```

**Outer Join/FUll Join**

it combines all the rows from both of table



```python
leftMerged=pd.merge(students,scores,on="ID",how="outer")
leftMerged
```

**Pandas Joins**

It is similar to our merge() but it will join the table based on index by default.

```python
df1=pd.DataFrame({
    "X":["X0","X1","X2"],
    "Y":["Y0","Y1","Y2"]
},index=["I0","I1","I2"])
df1
```

```python
df2=pd.DataFrame({
    "P":["P0","P1","P2"],
    "Q":["Q0","Q1","Q2"]
},index=["I0","I2","I3"])
df2
```

```python
joinOuterTable=df1.join(df2,how="outer")
joinOuterTable
```
