**Numpy**
* Find the average penguin body mass
* Extract the penguins which are heavier then the average body mass



```python
d1 = penguins["body_mass_g"].mean()
d1
```

```python
d1 = np.average(penguins["body_mass_g"])
d1
```

```python
penguins[penguins["body_mass_g"]>d1].head()
```

**Create a column(size category)**
* Classify the group into three categories
   * Light Weight:`<4000g`
   * mediumWeight:`4000g - 5000g`
   * heavyWeight:`>5000g`

```python
penguins[penguins["body_mass_g"]<4000].head()
```

```python
penguins[penguins["body_mass_g"] <= 5000].head()
```

```python
penguins[penguins["body_mass_g"]>5000].head()
```

```python
penguins["Size_Category"]=np.where(penguins["body_mass_g"]<4000,"LightWeight",
             np.where(penguins['body_mass_g']<=5000,"Medium","Heavy"))
penguins
```

```python
penguins[["species","body_mass_g","Size_Category"]]
```
