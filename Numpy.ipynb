##** Broadcasting**
1. If you add 1 dimensional array to 2D array,numpy will automatically expand the smaller array to the higher dimension array to match the shape

```python
a=np.array([[1,2,3],[4,5,6]])
b=np.array([7,8,9])
c=a+b
print(a)
print()
print(b)
print()
print(c)
```

## **Mathematical methods through numpy**

```python
a=np.array([1,2,3,4,5],ndmin=2)
#methods
print(f"Sum of all elements are: {np.sum(a)}")
print(f"Mean of the elements are: {np.mean(a)}")
print(f"Variance of the elements are: {np.var(a)}")
print(f"Standard Deviation of the elements are: {np.std(a)}")
print(f"Minimum of the elements are: {np.min(a)}")
print(f"Maximum of the elements are: {np.max(a)}")
print(f"product of all elments are: {np.prod(a)}")
print(f"Cummulative sum of all elements are: {np.cumsum(a)}")
```

**view() --> It makes a shallow copy of data**

```python
aa=np.array([1,2,3])
a=aa.view()
a[1] = 20
print(a)
print(aa)
```

**Copy() --> It will make an independent array**

```python
aa=np.array([1,2,3])
a=aa.copy()
a[1] = 20
print(a)
print(aa)
```

**hstack() --> It stacks arrays horizontally(side by side)**

```python
a=np.array([1,2,3])
b=np.array([5,6,7])
print(np.hstack((a,b)))
```

**vstack --> It stacks array vertically(top to bottom)**

```python
a=np.array([1,2,3])
b=np.array([5,6,7])
print(np.vstack((a,b)))
```

**Column_stack() --> stacks 1D array as columns into a 2D array**

```python
a=np.array([1,2,3])
b=np.array([5,6,7])
print(np.column_stack((a,b)))
```

**row_stack() --> stacks 1D array as rows**

**np.where() ==> It finds the indicies of elements that meets our condition**

```python
a=np.array([10,20,30,40,50,30])
#np.where(condition)
ans=np.where(a==30)
print(ans)
print(ans[0])
```

```python
#we are replacing all values that is less than 35 with 0 and others will remain as it is
n_a=np.where(a<35,0,a)
n_a
```

**How we will implement in 2D data**

**Elements that are greater than 3 ==> True --- rest will be false**

```python
a=np.array([[1,2,3],[7,8,9]])
ans=np.where(a>3,5,a)
print(f"Row Indices:{a[0]}")
print(f"Column Indices:{a[1]}")
```
