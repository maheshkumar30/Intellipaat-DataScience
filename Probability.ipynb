#**Probability**
It allow us to measure chances of an event to occur
* `p = number of favourable conditions / Total number of possible conditions`


**Marginal Probability**

It is a probability of single event happening without considering any other event
* Eg: In a bag we have 5 red balls and 3 blue balls (what is the marginal probabilty of picking red ball out of the bag)
* mp(r) = 5 red balls and total balls are 8 --> 5 / 8

**Joint Probability**

It is the probability of two or more than two event happening at the same time
* Q: What is the probabiltiy of two dice getting 3 at the same time
    * Sol: Each dice has 1/6 chances to get 3
      * The probabilty of getting 3 in dice is (1/6 * 1/6) = 1/36
      * total search space was 36 out of which 1 time we were getting (3,3)

**Conditional Probability**

It is a probability of an event happening given that the another event already happend
* Question: WHat is the probability of getting king when we know that the drawn cards are face cards from the deck
* Sol: There are total 12 face cards (King, Queen, Jack) * 4 suits
    * p = 4/12 --> 1/3

**Solution**

Total Outcome = {HH, HT, TH,TT}
* p(E) = {HT, TH} = 2/4 = 1/2
* p(F) = {HT, TH, HH} = 3/4
* p(E|F) = p(E n F)/p(F)
* p(EnF) = {HT,TH} = 2/4 = 1/2
* p(E|F) = (1/2) / (3/4) = 2/3

##**Probabilitic Distributions**
It tells us how likely different values of random variables are..

**Two types of probabilitic distribution**
###**1. Discrete Probability distribution**
**Bernoulli distribution**

* It is a distribution which experiments with only two possible outcomes
  * Success: 1
  * Failure: 0

It will have only two bars 0 and 1


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.choice([0,1], size = 1000, p= [0.3, 0.7])
```

```python
plt.hist(data,bins=2)
plt.show()
```

**Insight**
* Only two outcomes
* Height shows the probability of success and failure

**Binominal Distribution**

The number of success in a fixed number of trails (independent bernoulli trails)

* We can get discrete shape, based on probabilty value, data can be symmetric or skewed
* Example: Number of heads in 10 coin tosses

```python
binomData = np.random.binomial(n = 10, p = 0.2, size = 100000)
```

```python
import seaborn as sns
```

```python
sns.histplot(data = binomData, bins = 10)
plt.show()
```

**Insight**
* Values from 0 to number trails, frequency will be altered based on probability

**Poisson distribution**

It is a distribution that models the number of event occuring ina fixed interval of time
* We have one parameter : lambda (Average number of events)
* For small lambda -> We can get right skewed distribution
* As the lambda increase it becomes symmetric
* If the lambda is large-> we can get left skewed

**Examples**
* Number of projects we build per day
* Number of accidents per day

```python
poissionData = np.random.poisson(lam = 25, size = 10000)
poissionData
```

```python
plt.hist(poissionData,bins = 10)
plt.show()
```

**Insight**
* The center (25-30) we have more clusters

###**2. Continuous Probabilitic Distribution**

Continuous Distribution are used when values are measured and not counted

**Uniform Distribution**

In uniform distribution, all values have the equal range of data
* We can expect, flat block kind of distribution

**Example**
* Generating random number between any range
* Picking a uniform time between 8AM to 9AM

```python
uniformData = np.random.uniform(low = 0, high = 10, size = 10000)
uniformData[:5]
```

```python
plt.hist(uniformData, bins = 10)
plt.show()
```

**Insight**
* Flat distribution

`Note: Normal distribution is also a type of continuous probability distribution`
