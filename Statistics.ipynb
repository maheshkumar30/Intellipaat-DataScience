##**Variance**
It tells us how far each data points is from the mean

```python
data=[10,20,30,40]
variance=np.var(data,ddof=1)
print(variance)
```

* ddof - delta degree of freedom
* xBar -> Sample Mean
* u -> population mean
* x individual data points

###**Outlier Removal steps**
`**Note: Outlier can be found only in continuous numerical columns**`
* calculate Q! and Q3
* Find the IQR(Inter Quartile Range) Q3 - Q1
* We will find UpperFence and LowerFence
  * Upper Fence = Q3 + (1.5*IQR)
  * Lower Fence = Q1 - (1.5*IQR)
* We will check if there is any outlier present in my data(if any outlier ---> Remove)

```python
data = [10,12,13,15,33,26,70,40,45,50,52,-150,140]
a=np.percentile(data,25)
a #Q1
```

```python
b=np.percentile(data,75)
b #Q3
```

```python
#Upper Fence
UF=b + (1.5*(b-a))
UF
```

```python
#Lower Fence
LF=a - (1.5*(b-a))
LF
```

##**Correlation**
Relation between two variables,Two columns,Two values.It tells us how the feature are reated to each other.

**Note:** We can find correlation only for numerical data.
* Range of correlation varies from :-1 to +1
* Possitive correaltion:Between range of 0 to +1
* Negative correlation:Between range of -1 to 0
* Zero correlation or No correlation at 0


```python
#Postive corr -> (If x is inc ---> y will also increase)
"""
Example:
temp increase ---> Ice cream sale increases
"""
x = 10
y = x*2
print(x)
print(y)
```

```python
#Negative correlation (if x inc ---> y will Dec)
"""
Example:
Speed increae ----- Milage  decrease
"""
x = 5
y = 5/x
print(x)
print(y)
```

##**Kurtosis**
It help us measure taildness of a distribution
* Mesokurtic --> Normal
* Leptokurtic --> More tails
* Platykurtic --> Less tails

**Leptokurtic --> Postive kurtosis score**

```python
data = [10,11,11,12,12,12,12,12,12,12,12,12,13,13,14]
sb.kdeplot(data,fill = True)
plt.show()
```

```python
kurtosis(data)
```

**Platykurtic --> Negative Kurtosis score**

```python
data = [10,11,11,11,11,11,12,12,12,12,12,13,13,13,13,13,14,14,14,14,14,15]
sb.kdeplot(data, fill = True)
plt.show()
```
